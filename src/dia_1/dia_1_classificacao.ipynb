## Carregando os dados

Os dados estão sendo carregados, estes são alguns classificados, para se ter exemplos de como é a classificação e os dados sem classificação

```python
classificados = Carry.get_classificados()
sem_portes = Carry.get_no_carry()
```

```python
portes = []
for i in range(0, len(sem_portes)):
    lista = [sem_portes.loc[i]['altura'], sem_portes.loc[i]['peso']]
    portes.append(pd.DataFrame(Carry.get_distance(lista, pontos)).loc[1].argmin())
```

```python
classificados_porte = []
for i in range(0, len(portes)):
    classificados_porte.append(sem_portes.iloc[i].replace(np.nan, portes[i]))
```

```python
classificados_porte = pd.DataFrame(classificados_porte)
```
